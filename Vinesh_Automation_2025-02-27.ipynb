```python
# Rename the columns for each DataFrame

close.columns = ['Transaction Type', 'Identifier', 'Quantity']
resize.columns = ['Transaction Type', 'Identifier', 'Quantity']
new.columns = ['Transaction Type', 'Identifier', 'Quantity']
```

```python
# Concatenate the DataFrames vertically

trades = pd.concat([close, resize, new], ignore_index=True)
```

```python
trades.shape
```

```python
# Remove commas and convert Quantity to numeric while preserving NaNs
trades['Quantity'] = trades['Quantity'].replace(',', '', regex=True)
trades['Quantity'] = pd.to_numeric(trades['Quantity'], errors='coerce')

# Ensure the Quantity column is converted to integers
trades['Quantity'] = trades['Quantity'].astype(int)
```

```python
###

from openpyxl.utils import get_column_letter

# Define the output file name
output_file = r'CG_Order_Vinesh_2025-02-27.xlsx'

# Write the DataFrames to Excel
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    # Write the DataFrames without index
    Longs.to_excel(writer, sheet_name='Longs', index=False)
    Shorts.to_excel(writer, sheet_name='Shorts', index=False)

    # Access the workbook
    workbook = writer.book

    # Adjust column widths for each sheet
    for sheet_name in ['Longs', 'Shorts']:
        sheet = workbook[sheet_name]
        for col in sheet.columns:
            col_letter = get_column_letter(col[0].column)  # Get column letter
            sheet.column_dimensions[col_letter].width = 20

print(f"Excel file '{output_file}' has been created with all columns 20 cm wide.")
```
